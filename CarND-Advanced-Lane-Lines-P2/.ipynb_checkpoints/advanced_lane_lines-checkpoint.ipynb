## Camera calibration

```python
def calibrate_cam(calib_dir='./camera_cal/', verbose=False):
    
    if os.path.isfile(calib_dir+"calib_params.pickle"):
        file = open(calib_dir+"calib_params.pickle",'rb')
        ret, mtx, dist, rvecs, tvecs = pickle.load(file)
        file.close()
        
    else:
        # termination criteria
        criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.001)

        # prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
        objp = np.zeros((6*9,3), np.float32)
        objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

        # Arrays to store object points and image points from all the images.
        objpoints = [] # 3d point in real world space
        imgpoints = [] # 2d points in image plane.

        images = glob.glob(calib_dir+'calibration*.jpg')

        for fname in images:
            img = cv.imread(fname)
            gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

            # Find the chess board corners
            ret, corners = cv.findChessboardCorners(gray, (9,6), None)

            # If found, add object points, image points (after refining them)
            if ret == True:
                objpoints.append(objp)
                corners2 = cv.cornerSubPix(gray,corners, (11,11), (-1,-1), criteria)
                imgpoints.append(corners)

                if verbose:
                    # Draw and display the corners
                    cv.drawChessboardCorners(img, (9,6), corners2, ret)
                    cv.imshow('img', img)
                    cv.waitKey(500)

        if verbose:
            cv.destroyAllWindows()

        ret, mtx, dist, rvecs, tvecs = cv.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
        
        filehandler = open(calib_dir+"calib_params.pickle","wb")
        pickle.dump((ret, mtx, dist, rvecs, tvecs),filehandler)
        filehandler.close()
        
    return ret, mtx, dist, rvecs, tvecs
```

```python
ret, mtx, dist, rvecs, tvecs = calibrate_cam(verbose=True)
```

Now that we have the distortion coefficients, we can rectify images using cv.undistort(). Below are the images before and after the distortion correction.

```python
img     = cv.imread('./camera_cal/calibration1.jpg')
img_und = cv.undistort(img, mtx, dist, None, mtx)

# visualisation
f, ax = plt.subplots(1,2, constrained_layout=True, figsize=(20, 20))
ax[0].imshow(img)
ax[0].set_title('Distorted image', fontsize=20)
ax[1].imshow(img_und)
ax[1].set_title('Undistorted image', fontsize=20)
f.suptitle('Camera calibration and Distortion correction', fontsize=30, y=0.68)
```
